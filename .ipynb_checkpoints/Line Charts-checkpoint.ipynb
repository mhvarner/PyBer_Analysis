```python
import matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
savings=[0,100,150,200,300,500,300,350,500,600,750,1000]
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
```

```python
# x variables(where months is) goes on bottom y variables (where savings is) goes on right side which is why it is (months, savings)
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```
