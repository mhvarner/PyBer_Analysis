```python
%matplotlib inline
```

```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis=[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig,ax=plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Quick way: Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# plt function: Create the plot
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot and dd a label for the legend
plt.plot(x_axis,y_axis, label='Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a Legend
plt.legend()
```

```python
# Create the plot and dd a label for the legend
plt.plot(x_axis,y_axis, marker='*', color='blue',linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a grid
plt.grid()
# Add a Legend
plt.legend()
```

```python
# Create the plot and dd a label for the legend
plt.plot(x_axis,y_axis, marker='d', color='green',linewidth=2, label='Boston')
# Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Set the y limit between 0 and 45
plt.ylim(0,45)
# Create a title
plt.title('PyBer Fare by Month')
# Add a grid
plt.grid()
# Add a Legend
plt.legend()
```

```python
# BAR GRAPHS
# Create the plot
plt.bar(x_axis,y_axis)
```

```python
# Create the plot
plt.bar(x_axis,y_axis,color='green',label='Boston')
#Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis,y_axis)
```

```python
# Switch data to opposite side
# Create the plot
plt.barh(y_axis,x_axis)
```

```python
# Invert the y-axis to have months in ascending order
# Create the plot
plt.barh(x_axis,y_axis,color='magenta',label='Boston')
plt.gca().invert_yaxis()
#Create labels for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')
# Create a title
plt.title('PyBer Fare by Month')
# Add a legend
plt.legend()
```
